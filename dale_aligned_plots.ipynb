```python
# this function computes the accuracy of the overlapping plots

def compute_accuracy(pred, true):
    
    #pred is from 0 to 2.860596
    #true is from 0 to 247174
    
    #rescale pred values
    pred = (pred/2.860596)*247174
    pred = pred.astype(int)
    
    #now pred and true are on the same scale
    #compute proportion of overlap
    total_length =247174
    overlapping_length = 0
    def overlap_length(a,b,c,d):
        return min(b,d) - max(a,c)

    for i in range(len(true)-1):
        if i % 2 == 0:
            for j in range(0,len(pred)-2,2):
                if not (pred[j] > true[i+1] or pred[j+1] < true[i]):
                    overlapping_length = overlapping_length + overlap_length(true[i],true[i+1],pred[j],pred[j+1])
                    #print(overlapping_length)
        else:
            for j in range(1,len(pred)-2,2):
                if not (pred[j] > true[i+1] or pred[j+1] < true[i]):
                    overlapping_length = overlapping_length + overlap_length(true[i],true[i+1],pred[j],pred[j+1])
                    #print(overlapping_length)
                
    print("Accuracy is: " + str(round((overlapping_length/total_length)*100, 2)) + "% .")
```
